```
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
```

```
rcParams["figure.figsize"] = 8, 16
```

```
reader = easyocr.Reader(["en", "hi"])
```

```
from google.colab import files
uploaded = files.upload()
```

```
image = "pancard-500x500-500x500.jpg"
```

```
Image(image)
```

```
output = reader.readtext(image)
```

```
output
```

```
first_name_crdnt = output[5][0]
```

```
last_name_crdnt = output[6][0]
```

```
dob_crdnt = output[7][0]
```

```
pan_num_crdnt = output[9][0]
```

```
x_min, y_min = [int(min(idx)) for idx in zip(*first_name_crdnt,*last_name_crdnt,*dob_crdnt,*pan_num_crdnt)]
```

```
x_max, y_max = [int(max(idx)) for idx in zip(*first_name_crdnt,*last_name_crdnt,*dob_crdnt,*pan_num_crdnt)]
```

```
img = cv2.imread(image)
cv2.rectangle(img, (x_min, y_min), (x_max, y_max), (200, 0, 0),2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
